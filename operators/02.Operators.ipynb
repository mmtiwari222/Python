# Opeartors in Python

```python
# Arithmatic Operators
a = 10
b = 5
print("Addition:", a + b)        # Output: 15, + operator is used to add numbers
print("Subtraction:", a - b)     # Output: 5 , - operator is used to subtract numbers
print("Multiplication:", a * b)  # Output: 50 , * operator is used to multiply numbers
print("Division:", a / b)        # Output: 2.0, / operator is used to divide numbers (returns a float)
print("Floor Division:", a // b) # Output: 2, // operator is used to perform floor division (returns an integer)
print("Modulus:", a % b)        # Output: 0, % operator is used to get the remainder
print("Exponentiation:", a ** b) # Output: 100000, ** operator is used to raise a to the power of b
```

```python
# Relational/Comparison Operators
x = 10
y = 20
print("Equal to:", x == y)       # Output: False, == operator checks if values are equal
print("Not equal to:", x != y)   # Output: True, != operator checks if values are not equal
print("Greater than:", x > y)    # Output: False, > operator checks if x is greater than y
print("Less than:", x < y)       # Output: True, < operator checks if x is less than y
print("Greater than or equal to:", x >= y) # Output: False, >= operator checks if x is greater than or equal to y
print("Less than or equal to:", x <= y)    # Output: True, <= operator checks if x is less than or equal to y
```

```python
# Logical Operators
p = True # 1
q = False # 0
print("Logical AND:", p and q)  # Output: False, and operator returns True if both operands are True
# Truth table for AND:
# p     q     p and q
# True  True  True
# True  False False
# False True  False
# False False False


print("Logical OR:", p or q)    # Output: True, or operator returns True if at least one operand is True
# Truth table for OR:
# p     q     p or q
# True  True  True
# True  False True
# False True  True
# False False False


print("Logical NOT:", not p)   # Output: False, not operator returns the opposite of the operand
# Truth table for NOT:
# p     not p
# True  False
# False True
```

```python
# Bitwise Operators
m = 5  # In binary: 0101
n = 3  # In binary: 0011
print("Bitwise AND:", m & n)  # Output: 1 (In binary: 0001), & operator performs a bitwise AND operation
print("Bitwise OR:", m | n)   # Output: 7 (In binary: 0111), | operator performs a bitwise OR operation
print("Bitwise XOR:", m ^ n)  # Output: 6 (In binary: 0110), ^ operator performs a bitwise XOR operation
print("Bitwise NOT:", ~m)     # Output: -6 (In binary: 1010), ~ operator performs a bitwise NOT operation
print("Left Shift:", m << 1)  # Output: 10 (In binary: 1010), << operator shifts bits to the left ()
print("Right Shift:", m >> 1) # Output: 2 (In binary: 0010), >> operator shifts bits
```

```python
# Assignment Operators
x = 10
print("Initial value of x:", x) # Output: 10
x += 5
print("After x += 5:", x)       # Output: 15, x is updated to x + 5
x -= 3
print("After x -= 3:", x)       # Output: 12, x is updated to x - 3
x *= 2
print("After x *= 2:", x)       # Output: 24, x is updated to x * 2
x /= 4
print("After x /= 4:", x)       # Output: 6.0,  x is updated to x / 4 (returns a float)
x //= 2
print("After x //= 2:", x)      # Output: 3.0, x is updated to x // 2 (returns an integer)
x %= 2
print("After x %= 2:", x)       # Output: 1.0, x is updated to x % 2
x **= 3
print("After x **= 3:", x)      # Output: 1.0, x is updated to x ** 3
```

```python
# Membership Operators
my_list = [1, 2, 3, 4, 5]

# in operator checks if an element is present in a sequence (like list, tuple, string)
print("Is 3 in the list?", 3 in my_list)   # Output: True, in operator checks if an element is present in a sequence
print("Is 6 in the list?", 6 in my_list)   # Output: False, in operator checks if an element is present in a sequence

# not in operator checks if an element is not present in a sequence
print("Is 3 not in the list?", 3 not in my_list) # Output: False, not in operator checks if an element is not present in a sequence
print("Is 6 not in the list?", 6 not in my_list) # Output: True, not in operator checks if an element is not present in a sequence

```

```python
# Identity Operators
# is operator checks if two variables reference the same object in memory
# == operator checks if the values of two variables are equal (content of the objects)
a = [1, 2, 3]
b = a  # b references the same list as a
c = [1, 2, 3]  # c is a new list with
# the same content as a
print("a is b:", a is b)  # Output: True, is operator checks if two variables reference the same object 
print("a is c:", a is c)  # Output: False, is operator checks if two variables reference the same object
print("a == c:", a == c)  # Output: True, == operator checks if the values are equal (content of the lists)
```

```python
# Program - Find the sum of a three digit number entered by the user using op
number = int(input("Enter a three digit number: "))
# Extracting digits
hundreds = number // 100
tens = (number % 100) // 10
units = number % 10
# Calculating the sum of digits
digit_sum = hundreds + tens + units
print("The sum of the digits is:", digit_sum)
```
