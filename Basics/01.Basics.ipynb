# Type Conversion

```python
# Type Conversion in Python
# Type conversion is the process of converting a value from one data type to another.
# There are two types of type conversion:
# 1. Implicit Type Conversion (Type Coercion) 
# Implicit type conversion is the automatic conversion of one data type to another by the Python interpreter.

# Example of Implicit Type Conversion
a = 10
b = 4.5
c = a + b
print(c) # Output: 14.5
print(type(c)) # Output: <class 'float'>

# 2. Explicit Type Conversion (Type Casting)
# Explicit type conversion is the conversion of one data type to another by the programmer using built-in functions
# The functions used for type casting are:
# int() - converts a value to an integer
# float() - converts a value to a float
# str() - converts a value to a string
# bool() - converts a value to a boolean
# complex() - converts a value to a complex number

# Example of Explicit Type Conversion
a = 10
b = float(a)
print(b) # Output: 10.0
print(type(b)) # Output: <class 'float'>

c = str(a)
print(c) # Output: '10'
print(type(c)) # Output: <class 'str'>
```

# Literals

```python
# Literals in Python
# Literals are the values that are assigned to variables or used in expressions.
# There are several types of literals in Python:

# String Literals
username = "mmtiwari222"
print(username)

# Integer Literals
age = 23
print(age)

# Float Literals
weight = 70.5
print(weight)

# Boolean Literals
is_student = True
print(is_student)

# Complex Literals
z = 2 + 3j
print(z)
print(z.real, z.imag)

# Binary Literals
a = 0b1010
print(a)

# Octal Literals
b = 0o12
print(b)

# Hexadecimal Literals
c = 0xA
print(c)


```

# Take Input from User and Make a general calculator

```python
# Take input from user and make a simple calculator that takes two number as input and performs the four basic arithmetic operations (addition, subtraction, multiplication, division) on them and prints the results.

# Taking input from user

# 1st way - where we only take input from user without type conversion
num1 = input("Enter first number: ")
num2 = input("Enter second number: ")

print(type(num1), type(num2)) # Output: <class 'str'> <class 'str'>

print("Addition:", num1 + num2) # This will concatenate the two strings
# print("Subtraction:", num1 - num2) # This will give an error
# print("Multiplication:", num1 * num2) # This will give an error
# print("Division:", num1 / num2) # This will give an error


# 2nd way - where we take input from user and at the time of operation we convert the input to the required type
num1 = input("Enter first number: ")
num2 = input("Enter second number: ")

print("Addition:", int(num1) + int(num2))
print("Subtraction:", int(num1) - int(num2))
print("Multiplication:", int(num1) * int(num2))
print("Division:", int(num1) / int(num2))

print(type(num1), type(num2)) # Output: <class 'str'> <class 'str'>
# The type conversion functions (int(), float(), etc.) do not change the original variable, they return a new value of the specified type. Therefore, num1 and num2 remain as strings.

# 3rd way - where we take input from user and convert the input to the required type at the time of taking input
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

print("Addition:", num1 + num2)
print("Subtraction:", num1 - num2)
print("Multiplication:", num1 * num2)
print("Division:", num1 / num2)

print(type(num1), type(num2)) # Output: <class 'int'> <class 'int'>
```
